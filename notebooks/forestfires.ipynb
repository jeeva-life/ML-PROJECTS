```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Data Cleaning

```python
dataset[dataset.isnull().any(axis=1)]
```

### data conerting into 2 regions

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122:,"Region"] = 1
df = dataset
```

```python
df.head()
```

```python
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## as there are less number of NULL values, remove those records
```

```python
df = df.dropna().reset_index(drop = True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## remove it
```

```python
df = df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
## Now change the columns with respect to their column types
```

```python
df[['month','year','Temperature','RH','Ws']] = df[['month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
## conert respective columns into float datatypes
```

```python
object_column_names = [features for features in df.columns if df[features].dtypes == 'O']
```

```python
for i in object_column_names:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
# lets save the cleaned dataset
```

```python
df_copy = df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```
